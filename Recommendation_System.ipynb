```python

!pip install scikit-learn
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python


data = {
    'User': ['Alice', 'Bob', 'Charlie'],
    'Item1': [5, 3, 0],
    'Item2': [4, 0, 0],
    'Item3': [1, 1, 0],
    'Item4': [0, 1, 5],
    'Item5': [0, 0, 4],
}

df = pd.DataFrame(data)
df.set_index('User', inplace=True)
df

# Add Dave with some ratings
df.loc['Dave'] = [2, 0, 0, 0, 5]
df

```

```python
from scipy.stats import pearsonr

def pearson_similarity(df):
    n_users = df.shape[0]
    sim_matrix = np.zeros((n_users, n_users))
    for i in range(n_users):
        for j in range(n_users):
            if i != j:
                corr, _ = pearsonr(df.iloc[i], df.iloc[j])
                sim_matrix[i, j] = corr
            else:
                sim_matrix[i, j] = 1
    return sim_matrix

pearson_matrix = pearson_similarity(df)
pearson_df = pd.DataFrame(pearson_matrix, index=df.index, columns=df.index)
print(pearson_df)
```

```python

alice_similarities = pearson_df.loc['Alice'].drop('Alice')

most_similar_user = alice_similarities.idxmax()
print(f"User most similar to Alice: {most_similar_user}")


print(f"\n{most_similar_user}'s ratings:\n")
print(df.loc[most_similar_user])

print(f"\nAlice's ratings:\n")
print(df.loc['Alice'])


items_alice_hasnt_rated = df.loc['Alice'] == 0
recommendations = df.loc[most_similar_user][items_alice_hasnt_rated]

print("\nRecommended Items for Alice:")
print(recommendations[recommendations > 0])
```
